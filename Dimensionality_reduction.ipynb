### **Principal Component Analysis**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.spatial.transform import Rotation

m = 60
X = np.zeros((m,3))    # Initialize 3D dataset
np.random.seed(42)
angles = (np.random.rand(m) ** 3 + 0.5) * 2 * np.pi  # Uneven distribution
X[:, 0], X[:, 1] = np.cos(angles), np.sin(angles) * 0.5   # oval
X += 0.28 * np.random.randn(m, 3)   # add more noise
X = Rotation.from_rotvec([np.pi / 29, -np.pi / 20, np.pi / 4]).apply(X)
X += [0.2, 0, 0.2]
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)    #n_components --> select amount of data
X2D = pca.fit_transform(X)    # fit transform train data, but just transform test data
```

```python
X2D.shape
```

```python
pca.components_   #atiran xettlere component_ deyilir ve bunlar vektorlarin value-laridir.
```

```python
pca.explained_variance_ratio_    #Her kecirdiyi vektorda ne qeder data varsa faizle onu gosterir. Birinci kecdiyi vektorda 75%, ikinci ise 25%dir.
```

```python
1 - pca.explained_variance_ratio_.sum()
```

Ne qeder data itirmeye haziriqsa o qeder dimensionality reduction edirik

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False, parser = 'auto')
X_train, y_train, X_test, y_test = mnist.data[:60_000], mnist.target[:60_000], mnist.data[60_000:], mnist.target[60_000:]

pca = PCA()
pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
d
```

```python
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X_train)
```

```python
pca.n_components_
```

```python
%%time
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc.score(X_test, y_test)
```

```python
test_reduced = pca.transform(X_test)
```

```python
%%time

rfc.fit(X_reduced, y_train)
rfc.score(test_reduced, y_test)
```

```python
plt.figure(figsize=(10,7))
plt.plot(cumsum, linewidth=3)
plt.axis([0, 400, 0, 1])
plt.xlabel("Dimensions")
plt.ylabel("Explained Variance")
plt.plot([d, d], [0, 0.95], "k:")
plt.plot([0, d], [0.95, 0.95], "k:")
plt.plot(d, 0.95, "ko")
plt.annotate("Elbow", xy=(65, 0.85), xytext=(70, 0.7),
             arrowprops=dict(arrowstyle="->"))
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import   RandomizedSearchCV

pipeline = make_pipeline(PCA(), RandomForestClassifier(random_state= 42))

param_grid = {
    "pca__n_components": np.arange(10, 80),
    "randomforestclassifier__n_estimators": np.arange(50, 500)
}

rscv = RandomizedSearchCV(pipeline, param_grid, n_iter=10, cv=3)
rscv.fit(X_train[:1000], y_train[:1000])
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import   GridSearchCV

sgd_pipeline = make_pipeline(PCA(random_state= 42), SGDClassifier(random_state= 42))

sgd_param_grid = {
    "pca__n_components": np.arange(10, 80)
}

sgd_gscv = GridSearchCV(sgd_pipeline, sgd_param_grid, cv= 3)
sgd_gscv.fit(X_train[:10_00], y_train[:10_00])
```

Incremental PCA --> eger PCA edirikse butun data RAM a yigilir. Bu halda Incremental PCA kimi yollardan istifade edirik ki melumatlari hisse-hisse emal etsin.

### **PCA for compression**

```python
from sklearn.decomposition import IncrementalPCA

n_batches = 100
inc_pca = IncrementalPCA(n_components=154)
for X_batch in np.array_split(X_train, n_batches):
    inc_pca.partial_fit(X_batch)

X_reduced = inc_pca.transform(X_train)
```

### **LLM** - Local Linear Embedding

```python
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding

X_swiss, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10, random_state=42)
X_unrolled = lle.fit_transform(X_swiss)
```

```python
lle.reconstruction_error_
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Apply t-SNE to reduce dimensionality
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_swiss)

# Plot the results
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=t, cmap='Spectral')
plt.colorbar()
plt.title("Swiss Roll Unrolled with t-SNE")
plt.show()
```
